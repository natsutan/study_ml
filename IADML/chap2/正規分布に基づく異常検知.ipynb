## chap2 正規分布に従うデータからの異常検知

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
# データの読み込み
csv_file = '~/\myproj/study_ml/IADML/data/David.csv'
raw_data = pd.read_csv(csv_file)
raw_data
```

```python
# 身長と体重のデータのみ抜き出す。
data = raw_data.loc[:,['weight','height']]
data
```

```python
# 欠損データを落とす
# dropna()
# 身長、体重のデータには欠損データがないので省略

```

```python
# numpyへ変換
david = data.values
weight = david[:,0].astype(float)
height = david[:,1].astype(float)
```

```python
# データのプロット

plt.scatter(weight, height)
plt.xlabel('weight')
plt.ylabel('height')
plt.show()
```

```python
# ヒストグラムの表示
plt.hist(weight, bins=20)
plt.title('histgram of weight')
```

```python
# ヒストグラムの表示
plt.hist(height, bins=20)
plt.title('histgram of height')
```
