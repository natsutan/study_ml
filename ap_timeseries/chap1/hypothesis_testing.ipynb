```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df_historical = pd.read_csv('C:/home/myproj/study_ml/ap_timeseries/chap1/m_quote2.csv')
# df_historical
```

```python
df_historical.USD.plot()
```

```python
df_historical.USD.plot.hist()
```

```python
# 一次階差
df_diff1 = df_historical.USD.diff().dropna()
df_diff1.plot.hist()
```

```python
# p値の計算
# shapiroの 出力は(W統計量、p値)
import scipy.stats as stats

print("原系列の検定")
p = stats.shapiro(df_historical.USD.values)[1] * 100
print("p = ", p)
print("1次階差列の検定")
p = stats.shapiro(df_diff1.values)[1] * 100
print("p = ", p)
```
